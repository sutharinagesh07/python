# **7) EXPLAIN INDETAIL ABOUT VARIABLES WITH EXAMPLES AND RULES**
-   A Python variable is a reserved memory location to store values.
-   In other words, a variable in a python program gives data to the computer
     for processing.
-  Different data types in Python are Numbers, List, Tuple, Strings, Dictionary.

# **rules**
 Start with a letter or underscore
Variable names must start with a letter or an underscore, but not a number.

Use only certain characters
Variable names can only contain letters, numbers, and underscores (A-z, 0-9, and _).

Avoid spaces and special characters
Variable names cannot contain spaces or special characters like $, (, or *.

Avoid Python keywords
Avoid using Python keywords like "if", "for", or "while" as variable names.

Be case-sensitive
Variable names are case-sensitive, so "age", "Age", and "AGE" are three different variables.

Be descriptive
Variable names should be descriptive and not too short or too long.

Use snake_case
Use lowercase letters and underscores to separate words in variable names
